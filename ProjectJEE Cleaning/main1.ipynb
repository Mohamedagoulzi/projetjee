```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Amazon_Best_Seller_FINAL.csv')
print(df.shape)
df.head()
```

```python
print(df.columns)
print(df.dtypes)
```

```python
print("isnull :")
print(df.isnull().sum())
print("duplicated :" , df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print(df.shape)
```

```python
df.head()
```

```python
df['Price'] = df['Price'].str.replace('$', '')
df['Price'] = df['Price'].str.replace(',', '')
df['Price'] = df['Price'].astype(float)

df['Price'].head()
```

```python
df['Reviews Count'] = df['Reviews Count'].str.replace(',', '')
df['Reviews Count'] = df['Reviews Count'].astype(int)

df['Reviews Count'].head()
```

```python
df['No of Sellers'] = df['No of Sellers'].str.replace('Sellers', '')
df['No of Sellers'] = df['No of Sellers'].astype(int)

df['No of Sellers'].head()
```

```python
df['Rank'] = df['Rank'].str.replace('#', '')
df['Rank'] = df['Rank'].astype(int)

df['Rank'].head()
```

```python
print(df.dtypes)
print(df.isnull().sum())
```

```python
df['Price'] = df['Price'].fillna(df['Price'].median())
df['Reviews Count'] = df['Reviews Count'].fillna(df['Reviews Count'].median())
df['Rank'] = df['Rank'].fillna(df['Rank'].median())

print(df.isnull().sum())
```

```python
top_categories = df['Category'].value_counts().head(10).index

df['Category_grouped'] = df['Category'].apply(
    lambda x: x if x in top_categories else 'Other'
)

df[['Category', 'Category_grouped']].head(10)
```

```python
encoded = pd.get_dummies(df['Category_grouped'])
encoded.head()
```

```python
df = pd.concat([df, encoded], axis=1)

df.head()
```

```python
df = df.replace({True: 1, False: 0})
df.head()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['Price'])
plt.title("Bote  moustache - Valeurs aberrantes (Price)")
plt.ylabel("Price")
plt.show()
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Limite basse =", lower_limit)
print("Limite haute =", upper_limit)
```

```python
df['Price'] = df['Price'].apply(
    lambda x: x if (x >= lower_limit and x <= upper_limit) else None
)

df['Price'].isnull().sum()
```

```python
median_price = df['Price'].median()
df['Price'] = df['Price'].fillna(median_price)

df['Price'].isnull().sum()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['Price'])
plt.title("Bote  moustache - Valeurs aberrantes (Price)")
plt.ylabel("Price")
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import random

session = requests.Session()  # Faster than calling requests.get each time

def readProductTitle(url):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36",
            "Accept-Language": "en-US,en;q=0.9"
        }
        
        # timeout prevents hanging forever
        response = session.get(url, headers=headers, timeout=5)

        soup = BeautifulSoup(response.text, "html.parser")
        title = soup.find("span", id="productTitle")

        # small delay to avoid Amazon blocking
        sleep(random.uniform(0.5, 1.2))

        return title.get_text().strip() if title else "Title not found"

    except Exception as e:
        return "Error"

df['Product Title'] = df['Product Link'].apply(readProductTitle)
```
